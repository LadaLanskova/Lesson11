```python
'''
f(x) = -12x^4*sin(cos(x)) - 18x^3+5x^2 + 10x - 30
Определить корни
Найти интервалы, на которых функция возрастает
Найти интервалы, на которых функция убывает
Построить график
Вычислить вершину
Определить промежутки, на котором f > 0
Определить промежутки, на котором f < 0
Выполните все пункты для диапазона по оси X = [-100:100]
'''
```

```python
import sympy
import matplotlib
import math
import scipy
```

```python
x = sympy.Symbol('x')
```

```python
f_x = - 12 * x ** 4 * sympy.sin(sympy.cos(x)) - 18 * x **3 + 5 * x ** 2 + 10 * x - 30
print(f_x)
```

```python
for i in range (0, 51):
    print(i/10, f_x.subs(x, i/10).evalf())
```

```python
print(f_x.subs(x, 2.27305649))
print('первый корень ~', 2.27305649)
```

```python
print(f_x.subs(x, 4.38352367))
print('второй корень ~', 4.38352367)
```

```python
diff_f_x = f_x.diff(x)
print(diff_f_x.subs(x, 3.8193108))
print('экстремум', 3.8193108)
```

```python
sympy.plotting.plot(f_x, (x, 0, 5))
```

```python
sympy.plotting.plot(diff_f_x, (x, 3.819, 3.82))
```

```python
last = -5
for i in range (0, 101):
    tmp = f_x.subs(x, i/10).evalf()
    if last * tmp <= 0:
        print(i/10, f_x.subs(x, i/10).evalf())
    last = tmp
print ('функция в диапазоне х от 0 до 10 положительна в интервале х от ~ 2.3 до ~ 4.3 и от ~ 8.1 до 10')
```

```python
last = -5
for i in range (0, 51):
    tmp = diff_f_x.subs(x, i/10).evalf()
    if last * tmp <= 0:
        print(i/10, diff_f_x.subs(x, i/10).evalf())
    last = tmp
print ('функция в диапазоне х от 0 до 5 возрастает в интервале х от 0 до ~ 0.5 и от ~ 1.8 до ~ 3.8')
```
